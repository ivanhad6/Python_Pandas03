```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows =10
n_cols=5

cols = tuple('ABCDE')

df=pd.DataFrame(np.random.randint(1,20, size=(n_rows,n_cols)), columns = cols)

df
```

```python
#Cara melakukan seleksi kolom dan baris menggunakan loc

df.loc[[0,3,4], ['B','E']]
```

```python
#Cara seleksi baris dengan kondisi

df.loc[df['B']>10,['B','D','E']]
```

```python
#Cara melakukan slicing Data Frame dengan loc

df.loc[0:4,'B':'D']
```
